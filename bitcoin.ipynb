```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, GRU, Bidirectional, Flatten
from sklearn.model_selection import train_test_split
```

```python
datapath = '/content/drive/MyDrive/Colab Notebooks/dataset/Bitcoin/gemini_BTCUSD_day.csv'
pd.options.display.float_format = '{:,.3f}'.format
data = pd.read_csv(datapath, usecols=['Open',	'High',	'Low',	'Close'], header=0, skiprows=1).values
```

```python

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
prediction_days = 30

X_train = X_test = X_val = []
y_train = y_test = y_val = []
```

```python
X = []
y = []
for x in range(prediction_days, len(scaled_data)):
  X.append(scaled_data[x - prediction_days: x, :])
  y.append(scaled_data[x, :])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle = False)
X_train, X_val, y_train, y_val =  train_test_split(X_train, y_train, test_size=0.2, shuffle = False)
```

```python
print(len(y_train))
print(len(y_test))
print(len(y_val))
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
X_test, y_test = np.array(X_train), np.array(y_train)
X_val, y_val = np.array(X_train), np.array(y_train)
```

## GRU

```python
model = Sequential()
model.add(Bidirectional(GRU(units=30, return_sequences=True, dropout=0.1), input_shape=(X_train.shape[1:])))
model.add(Bidirectional(GRU(units=60, return_sequences=True, dropout=0.1)))
model.add(Dense(500, activation="relu"))
model.add(Flatten())
model.add(Dense(1000, activation="relu"))
model.add(Dense(100, activation="relu"))
model.add(Dense(4, activation="softmax"))
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), batch_size=32, epochs=25)
```

```python
loss, acc = model.evaluate(X_test, y_test, verbose=1)
print('Restored model, accuracy: {:5.2f}%'.format(100*acc))
```
